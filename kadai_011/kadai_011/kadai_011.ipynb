```python
# 惑星のリスト
planets = ["水","金","地","火","木","土","天","海","冥"]

# for文を使った繰り返し
print("【for文】")
for planet in planets:
    print(planet)

for _ in range(2):
    for planet in planets:
       print(planet)
```

```python
# 惑星のリスト
planets = ["水","金","地","火","木","土","天","海","冥"]

# while文を使った繰り返し
print("【while文】")

count = 0
while count < 2:
  for planet in  planets:
      print(planet)
  count += 1
```
